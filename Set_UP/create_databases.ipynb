```python
# Delta Lake test karo
# Unity Catalog ke andar table banate hain

spark.sql("USE CATALOG financial_market")
spark.sql("USE DATABASE bronze")

test_data = [(1, "stocks"), (2, "crypto"), (3, "forex")]
test_df = spark.createDataFrame(test_data, ["id", "asset_type"])

test_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("financial_market.bronze.test_table")

print("✅ Delta Lake working correctly.")

# verify
spark.sql("SELECT * FROM financial_market.bronze.test_table").show()

# cleanup
spark.sql("DROP TABLE IF EXISTS financial_market.bronze.test_table")
print("✅ Test table cleaned up.")
print("\n🚀 Setup complete. Ready to start ingestion.")
```
