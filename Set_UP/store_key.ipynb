```python
# yeh cell safe hai — key values nahi hain
def get_api_keys():
    """
    Returns API keys from environment variables.
    Always run 02_config notebook first.
    """
    alpha_key    = os.environ.get("ALPHA_VANTAGE_API_KEY")
    exchange_key = os.environ.get("EXCHANGERATE_API_KEY")

    if not alpha_key or not exchange_key:
        raise ValueError("❌ Keys not loaded. Run 02_config first.")

    return alpha_key, exchange_key

# test karo
alpha_key, exchange_key = get_api_keys()
print(f"Alpha key: {alpha_key[:4]}****")
print(f"Exchange key: {exchange_key[:4]}****")
print("✅ Config ready")
```
