```python
# ============================================
# SECRETS SETUP - ONE-TIME CONFIGURATION
# ============================================
# Run this cell ONCE to store your API keys

import requests

workspace_url = spark.conf.get("spark.databricks.workspaceUrl")
token = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()

# Create scope
try:
    requests.post(
        f"https://{workspace_url}/api/2.0/secrets/scopes/create",
        headers={"Authorization": f"Bearer {token}"},
        json={"scope": "fin-project"}
    )
    print("✅ Scope created")
except:
    print("ℹ️ Scope already exists")

# Store Alpha Vantage key
response1 = requests.post(
    f"https://{workspace_url}/api/2.0/secrets/put",
    headers={"Authorization": f"Bearer {token}"},
    json={
        "scope": "fin-project",
        "key": "alpha_vantage_key",
        "string_value": "key_here" 
    }
)

# Store ExchangeRate key
response2 = requests.post(
    f"https://{workspace_url}/api/2.0/secrets/put",
    headers={"Authorization": f"Bearer {token}"},
    json={
        "scope": "fin-project",
        "key": "exchangerate_key",
        "string_value": "key_here"  
    }
)

if response1.status_code == 200:
    print("✅ Alpha Vantage key stored")
else:
    print(f"❌ Error: {response1.text}")

if response2.status_code == 200:
    print("✅ ExchangeRate key stored")
else:
    print(f"❌ Error: {response2.text}")
```

```python
def get_api_keys():
    """
    Returns API keys from Databricks Secrets.
    Secrets must be created first (run Cell 1).
    """
    try:
        alpha_key = dbutils.secrets.get(scope="fin-project", key="alpha_vantage_key")
        exchange_key = dbutils.secrets.get(scope="fin-project", key="exchangerate_key")
        
        if not alpha_key or not exchange_key:
            raise ValueError("❌ Keys not found in secrets")
        
        return alpha_key, exchange_key
    
    except Exception as e:
        raise ValueError(f"❌ Error loading keys: {e}\nRun Cell 1 to create secrets first.")

# Test karo
alpha_key, exchange_key = get_api_keys()
print(f"Alpha key: {alpha_key[:4]}****")
print(f"Exchange key: {exchange_key[:4]}****")
print("✅ Config ready")
```
